# Utilities

```python
def convert_to_array(string):
    string = string.replace("\n", "").replace("[", "").replace("]", "").replace(" ", "").split(",")
    result = []
    for item in string:
        result.append(float(item))
    return np.array(result)
```

# Data Processing

```python
data = pd.read_csv('www.daraz.pk_QSS_test.csv', index_col=0)
```

```python
data.head()
```

```python
data['AT_TAGS'] = data['AT_TAGS'].apply(lambda x: "invalid_event" if "invalid_event" in x else "valid_event")
```

```python
data.head()
```

```python
for i in range(5, 10):
    data.iloc[:, i] = data.iloc[:, i].apply(lambda x: convert_to_array(x).mean())
```

```python
data.head()
```

```python
valid = data[data['AT_TAGS'] == "valid_event"]
invalid = data[data['AT_TAGS'] == "invalid_event"]
```

```python
valid_features = valid.iloc[:, [0, 5, 6, 7, 8, 9, 10, 11, 12]]
invalid_features = invalid.iloc[:, [0, 5, 6, 7, 8, 9, 10, 11, 12]]
```

```python
vcolumns = []
invcolumns = []
columns = []
for i in range(9):
    col = []
    temp = valid_features.iloc[:, i].to_numpy()
    temp = temp/max(temp)
    vcolumns.append(temp)
    temp = invalid_features.iloc[:, i].to_numpy()
    temp = temp/max(temp)
    invcolumns.append(temp)
```

```python
plt.figure(figsize=(20, 10))
bp0 = plt.boxplot(vcolumns, showfliers=False, patch_artist=True)
bp1 = plt.boxplot(invcolumns, showfliers=False, patch_artist=True)
for box in bp0['boxes']:
    box.set(color='blue', linewidth=8)
    box.set(facecolor='white')
for box in bp1['boxes']:
    box.set(color='red', linewidth=1)
    box.set(facecolor='white')
plt.legend([bp0['boxes'][0], bp1['boxes'][0]], ['valid', 'invalid'], loc='upper left')
plt.xticks([1,  2,  3,  4,  5,  6,  7,  8,  9], 
           ['acceleration', 'speed', 'gps_speed',
           'AX', 'AY', 'AZ', 'GRAVX', 'GRAVY',
           'GRAVZ'])
plt.show()
```

```python
vf = valid_features.iloc[:, [2, 3, 5]]
invf = invalid_features.iloc[:, [2, 3, 5]]
print(len(vf))
print(len(invf))
```

```python
def remove_outliers(data):
    df = data
    # Select the first quantile
    for i in df.columns:
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1

        # Filtering Values between Q1-1.5IQR and Q3+1.5IQR
        df = df.query(f'(@Q1 - 1.5 * @IQR) <= {i} <= (@Q3 + 1.5 * @IQR)')
    return df
```

```python
vf = remove_outliers(vf)
invf = remove_outliers(invf)
print(len(vf))
print(len(invf))
```
